### Логистическая регрессия

1. Загрузите данные `load_wine` из `sklearn.datasets`. Из данных исключите объекты класса 2.
 Отмасштабируйте признаки, используя класс `StandardScaler` с гиперпараметрами по умолчанию. Обучите логистическую регрессию и оцените важность признаков. Укажите название признака, который оказался наименее значимым.

Обратите внимание, целевое значение лежит по ключу `'target'`, матрица объекты-признаки лежит по ключу `'data'`



```python
import numpy as np
from sklearn.datasets import load_wine
```

Загружаем датасет с винами и «складываем» объекты-признаки и целевые значения в отдельные переменные.

```python
wine = load_wine()
data, target = wine['data'], wine['target']
data, target
```

Находим индексы объектов класса 2 (по целевым значениям).

```python
indices_of_class_two = np.where(target == 2)
indices_of_class_two
```

Убираем объекты класса 2.

```python
data = np.delete(data, indices_of_class_two, axis=0)
target = np.delete(target, indices_of_class_two, axis=0)
data, target
```

Масштабируем признаки, используя `sklearn.preprocessing.StandardScaler` с параметрами по-умолчанию.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(data)
data = scaler.transform(data)
data
```

Делим датасет на обучающую и тестовую выборки.

```python
from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(data, target)
data_train, data_test, target_train, target_test
```

Обучаем `sklearn.linear_model.LogisticRegression` данным из обучающей выборки и измеряем точность полученной модели на тестовой.

```python
from sklearn.linear_model import LogisticRegression

estimator = LogisticRegression()
estimator.fit(data_train, target_train)
f'{estimator.score(data_test, target_test):.2%}'
```

Находим индекс наименее значимого признака.

```python
np.argmin(estimator.coef_)
```

4. Загрузите файл SMSSpamCollection из UCI (https://archive.ics.uci.edu/ml/machine-learning-databases/00228/). Данные содержат текстовую информацию и бинарное целевое значение (‘spam’, ‘ham’), Пусть в обучающую часть попадут первые 4000 объектов из таблицы, в тестовую часть оставшиеся объекты. Обучите `TfidfVectorizer` с гиперпараметрами по умолчанию на текстах из обучающей части и получите векторное представление для объектов обучающей и тестовой части. Укажите полученное число признаков.

```python
import pandas as pd
```

Загружаем датасет с SMS как CSV-файл с разделитем-табом. Называем колонку с целевым значения (классом сообщения) `target`, а колонку с текстом сообщения `data`.

```python
sms = pd.read_csv('data/sms/SMSSpamCollection', delimiter='\t', names=['target', 'data'])
sms
```

Разделям датасет на обучающую и тестовую выборки.

```python
sms_train = sms[:4000]
sms_test = sms[4001:]
```

Получим векторное представление текстов сообщений из обучающей и тестовой выборки с помощью `sklearn.feature_extraction.text.TfidfVectorizer` с параметрами по-умолчанию.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectorizer.fit(sms_train['data'])

vectorizer.transform(sms_train['data'])
vectorizer.transform(sms_test['data'])

len(vectorizer.get_feature_names())
```
